```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
U_14 = np.array([-1, -0.5, 0, 0.3, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85, 0.9, 0.95, 1, 1.3, 1.5, 1.7])
U_12 = np.array([-1, -0.5, 0, 0.3, 0.48, 0.51, 0.54, 0.56, 0.58, 0.59, 0.60, 0.62, 0.63, 0.64, 0.66, 0.7, 0.72, 0.74])

I = (U_14 - U_12)/100

# I = [0, 10, 20, 60, 100, 180, 240, 320, 400, 500]
U = U_12

# Исходные точки
x = np.array(U)
y = np.array(I)

# Создаем больше точек для сглаживания
x_new = np.linspace(x.min(), x.max(), 300)  # Больше точек для сглаживания
spl = make_interp_spline(x, y, k=2)  # k=3 задает кубический сплайн
y_smooth = spl(x_new)

# Построение графика
plt.plot(x_new, y_smooth, label="Сглаженный график")
plt.scatter(x, y, color='red', label="Исходные точки")
plt.title('ВАХ кремниевого диода')
plt.xlabel('Uсм, B')
plt.ylabel('I, мА')
plt.legend()
plt.show()
```

```python
R = np.array([100, 240, 510, 750, 1100, 2400, 4700, 9100, 16000, 33000])
tau = np.array([10, 9, 8, 8, 7, 6.75, 6.5, 6, 5.5, 5])
teta = tau/23/2
R_i = []
for i in range(3):
    Ri = R[i]*(np.tan(teta[i])-teta[i])/np.pi
    R_i.append(Ri)
# Исходные точки
x = np.array(R)
y = np.array(teta)

# Создаем больше точек для сглаживания
x_new = np.linspace(x.min(), x.max(), 300)  # Больше точек для сглаживания
spl = make_interp_spline(x, y, k=1)  # k=3 задает кубический сплайн
y_smooth = spl(x_new)

# Построение графика
plt.plot(x_new, y_smooth, label="Сглаженный график")
plt.scatter(x, y, color='red', label="Исходные точки")
plt.title('Зависимость угла отсечки от Rн')
plt.xlabel('Rн, кОм')
plt.ylabel('θ')
plt.legend()
plt.show()
```
