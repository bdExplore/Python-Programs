```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
I = np.array([0, 0.4, 1, 1.5, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.2, 7.3, 7.4])
U = np.array([-1.95, -1.8, -1.6, -1.5, -1.4, -1.35, -1.25, -1.15, -1.1, -1, -0.95, -0.85, -0.76, -0.65, -0.45, -0.25, -0.1, 0])

# Исходные точки
x = np.array(U)
y = np.array(I)

# Создаем больше точек для сглаживания
x_new = np.linspace(x.min(), x.max(), 300)  # Больше точек для сглаживания
spl = make_interp_spline(x, y, k=2)  # k=3 задает кубический сплайн
y_smooth = spl(x_new)

# Построение графика
plt.plot(x_new, y_smooth, label="Сглаженный график")
plt.scatter(x, y, color='red', label="Исходные точки")
plt.title('')
plt.xlabel('U=, B')
plt.ylabel('I=, мА')
plt.legend()
plt.show()
```
