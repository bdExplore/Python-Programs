```python
def find_max_bouquet_cost(n, budgets):
    from itertools import combinations

    MAX_POWER = 100
    pow_of_two = [2 ** i for i in range(MAX_POWER + 1)]

    three_color_sums = set()
    for combo in combinations(pow_of_two, 3):
        three_color_sums.add(sum(combo))

    sorted_sums = sorted(three_color_sums)

    results = []
    for budget in budgets:
        left, right = 0, len(sorted_sums) - 1
        max_cost = -1
        while left <= right:
            mid = (left + right) // 2
            if sorted_sums[mid] <= budget:
                max_cost = sorted_sums[mid]
                left = mid + 1
            else:
                right = mid - 1
        results.append(max_cost if max_cost != -1 else -1)

    return results


n = int(input())
budgets = [int(input()) for _ in range(n)]

results = find_max_bouquet_cost(n, budgets)
for res in results:
    print(res)
```

```python
def min_adj(n, m, a):
    def good_days_with_changes(max_changes):
        
        modified = a[:]
        good_days = 0
        changes = 0

        for i in range(2, n):
            if a[0] <= modified[i] <= a[1]:
                good_days += 1
            else:
                if modified[i] < a[0]:
                    changes_needed = a[0] - modified[i]
                    if changes + changes_needed <= max_changes:
                        changes += changes_needed
                        modified[i] = a[0]
                        good_days += 1
                elif modified[i] > a[1]:
                    changes_needed = modified[i] - a[1]
                    if changes + changes_needed <= max_changes:
                        changes += changes_needed
                        modified[i] = a[1]
                        good_days += 1

            if good_days >= m:
                return True

        return good_days >= m

    left, right = 0, max(a) - min(a)
    result = right

    while left <= right:
        mid = (left + right) // 2
        if good_days_with_changes(mid):
            result = mid
            right = mid - 1
        else:
            left = mid + 1

    return result

n, m = map(int, input().split())
a = list(map(int, input().split()))

print(min_adj(n, m, a))
```

```python
def min_operations(n, x, y, z, a):
    from math import ceil

    def find_steps_to_multiple(value, divisor):
        
        remainder = value % divisor
        return 0 if remainder == 0 else divisor - remainder

    operations = 0

    while True:
        
        steps = []
        for ai in a:
            steps.append((find_steps_to_multiple(ai, x), find_steps_to_multiple(ai, y), find_steps_to_multiple(ai, z)))

        
        x_covered = any(ai % x == 0 for ai in a)
        y_covered = any(ai % y == 0 for ai in a)
        z_covered = any(ai % z == 0 for ai in a)

        if x_covered and y_covered and z_covered:
            break  


        best_option = (float('inf'), -1, -1) 

        for i in range(n):
            for j, divisor in enumerate([x, y, z]):
                if not [x_covered, y_covered, z_covered][j]:
                    if steps[i][j] < best_option[0]:
                        best_option = (steps[i][j], i, j)

        
        min_steps, idx, condition_idx = best_option
        a[idx] += min_steps
        operations += min_steps

    return operations

if __name__ == "__main__":
    n, x, y, z = map(int, input().split())
    a = list(map(int, input().split()))
    
    print(min_operations(n, x, y, z, a))
```

```python
def is_non_degenerate(x1, y1, x2, y2, x3, y3):
    return (x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) != 0

def max_happy_triples(points):
    n = len(points)
    if n % 3 != 0:
        return n // 3
    
    used = [False] * n  
    count = 0
    
    for i in range(n):
        if used[i]:  
            continue
        for j in range(i + 1, n):
            if used[j]:  
                continue
            for k in range(j + 1, n):
                if used[k]:  
                    continue
                x1, y1 = points[i]
                x2, y2 = points[j]
                x3, y3 = points[k]
                if is_non_degenerate(x1, y1, x2, y2, x3, y3):
               
                    used[i] = used[j] = used[k] = True
                    count += 1
                    break  
        if count == n // 3:  
            break
            
    return count


n = int(input())
points = [tuple(map(int, input().split())) for _ in range(n)]

print(max_happy_triples(points))
```

```python
MOD = 998244353

def f(n, k, a):
    pair_sums = []
    for i in range(n):
        for j in range(i + 1, n):
            pair_sums.append(a[i] + a[j])
    
    results = []
    for p in range(1, k + 1):
        total = sum(pow(x, p, MOD) for x in pair_sums) % MOD
        results.append(total)
    
    return results

n, k = map(int, input().split())
a = list(map(int, input().split()))

results = f(n, k, a)

for res in results:
    print(res)
```
