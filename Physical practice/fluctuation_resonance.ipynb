```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
```

```python
I = [0, 10, 20, 60, 100, 180, 240, 320, 400, 500]
U = [0.4, 0.46, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8, 0.85]

# Исходные точки
x = np.array(U)
y = np.array(I)

# Создаем больше точек для сглаживания
x_new = np.linspace(x.min(), x.max(), 300)  # Больше точек для сглаживания
spl = make_interp_spline(x, y, k=3)  # k=3 задает кубический сплайн
y_smooth = spl(x_new)

# Построение графика
plt.plot(x_new, y_smooth, label="Сглаженный график")
plt.scatter(x, y, color='red', label="Исходные точки")
plt.title('ВАХ диода')
plt.xlabel('Uсм, B')
plt.ylabel('I, мА')
plt.legend()
plt.show()
```

```python
Uf_rez = [0.1, 0.2, 0.31, 0.4, 0.51, 0.6, 0.7, 0.81, 0.9, 1.0]
f_rez = [817.5, 841, 847.3, 863.4, 872.8, 883.2, 885.5, 906.7, 912.1, 925.9]
# Исходные точки
x = np.array(Uf_rez)
y = np.array(Uf_rez)

ff_rez = np.array(f_rez) * np.array(f_rez)
C = 1/(4*np.pi*np.pi*0.28*ff_rez)
y = C
# Создаем больше точек для сглаживания
x_new = np.linspace(x.min(), x.max(), 300)  # Больше точек для сглаживания
spl = make_interp_spline(x, y, k=1)  # k=3 задает кубический сплайн
y_smooth = spl(x_new)

# Построение графика
plt.plot(x_new, y_smooth, label="Сглаженный график")
plt.scatter(x, y, color='red', label="Исходные точки")
plt.title('Вольтфарадная хар-ка диода')
plt.xlabel('Uсм, B')
plt.ylabel('C, Ф')
plt.legend()
plt.show()
```
